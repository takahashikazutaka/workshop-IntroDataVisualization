```python
import numpy as np
import matplotlib.pyplot as plt 
import seaborn
# We activate the inline mode in the Notebook 
% matplotlib inline
y = np.random.randn(1000)
# Plot the newly generated random vector with 1000 data points against the index 1:1000 
plt.plot(y)
# Saving a plot as a file is very easy! 
plt.savefig('randomplot.pdf')
```

```python
# We can plot the graph of a mathematical function 
x = np.linspace(-10,10,1000)
y = np.sin(3*x)*np.exp(-0.1*x**2)
plt.plot(x,y)
```

```python
# Specifications of the plot
# We can plot the graph of a mathematical function 
x = np.linspace(-10,10,1000)
y = np.sin(3*x)*np.exp(-0.1*x**2)
plt.plot(x,y,'--^',lw=3,color='#fdbb82',mfc='#2b8cbe',ms=8)
# where 
# The 3rd argument is the line style: 
#   "--" indicates a dashed line 
#   "^" indicates an upper triangle marker 
# lw argument is line width 
# color argument is for line color in hex RGB value 
# mfc argument is for marker face color 
# ms argument indicates the marker size 
```

```python
# Scatter plot example 
x = np.random.randn(100)
y = x + np.random.randn(100)
plt.scatter(x,y)
```

```python
# Now we combine them into two subplots in a row and convert the second plot into a polar plot 
# Left panel 
plt.subplot(1,2,1)
x = np.linspace(-10,10,1000)
y1 = np.sin(3*x)*np.exp(-0.1*x**2)
plt.plot(x,y1,'--^',lw=1,color='navy',mfc='#2b8cbe',ms=8,label='sin(3x)*exp(x^2)')
y2 = np.cos(3*x)
plt.plot(x,y2,'--*',lw=1,color='yellow',mfc='#1b8cbe',ms=8,label='cos(3x)')

plt.xticks([-10,0,10])
plt.yticks([-1,0,1])
# Since we are going to put legends and we know that functions are bounded by \pm 1, we will expand the vertical range 
# so that we can account for some space for the legend for two line plots 
plt.ylim(-2,2)
plt.xlabel("X axis label")
plt.ylabel("Y axis label")
plt.legend()

# Right panel 
plt.subplot(1,2,2, polar=True)
x = np.linspace(0,2*np.pi,1000) # 2*np.pi*np.random.randn(1000)
plt.plot(x,1+2*np.cos(6*x))

# Now that the plot is in polar coordiante, the tick for the x can also be in polar 
plt.xticks(np.linspace(0,5*np.pi/3,6))
plt.title('A polar plot')

```

```python
# Simple animation to draw a plot usin nbagg backend 
%matplotlib nbagg
import matplotlib.pyplot as plt
import matplotlib.animation
import numpy as np

t = np.linspace(0,2*np.pi)
x = np.sin(t)

fig, ax = plt.subplots()
l, = ax.plot([0,2*np.pi],[-1,1])

animate = lambda i: l.set_data(t[:i], x[:i])

ani = matplotlib.animation.FuncAnimation(fig, animate, frames=len(t))

plt.show()
```
